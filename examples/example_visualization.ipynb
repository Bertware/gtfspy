## First just some necessary imports: 

```python
import sys
sys.path.append("../") 
# To be able to import the current version of gtfspy

from gtfspy import mapviz
from example_import import load_or_import_example_gtfs
from matplotlib import pyplot as plt
```

## Simple route and stop visualizations

```python
g = load_or_import_example_gtfs()
# g is now a gtfspy.gtfs.GTFS object

# Plot the route network and all stops to the same axes
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax = mapviz.plot_route_network_from_gtfs(g, scalebar=True, ax=ax, map_style="light_all")
mapviz.plot_all_stops(g, ax, map_style="light_all") # Remember to pass the map style here as well!

plt.show()
```

## Thumbnail figures highlighting central areas:

```python
# Plot also a thumbnail figure highlighting the central areas:
ax_thumbnail = mapviz.plot_route_network_thumbnail(g, map_style="dark_all")
# For saving the figure: ax_thumbnail.figure.savefig("test_thumbnail.jpg")
plt.show()
```
